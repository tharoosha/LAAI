```python
import json
train_dataset_path = "../combined.json"
val_dataset_path = "../eval.json"
with open(val_dataset_path, 'r') as f:
    try:
        data = json.load(f)
        for i, entry in enumerate(data):
            if not all(k in entry for k in ["context", "user_input", "intent"]):
                print(f"Missing field in entry {i}: {entry}")
            elif not isinstance(entry['context'], str) or not isinstance(entry['user_input'], str):
                print(f"Type issue in entry {i}: {entry}")
    except json.JSONDecodeError as e:
        print(f"JSON parsing error: {e}")
```

```python
# Import necessary libraries
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, Trainer
from datasets import load_dataset
from peft import LoraConfig, get_peft_model, TaskType
from transformers import DataCollatorForSeq2Seq
import numpy as np

# Define model parameters
model_id = "microsoft/Phi-3-mini-4k-instruct"
tokenizer_id = model_id
# dataset_path = "/content/drive/MyDrive/intent-classifier/data/combined.json"
output_dir = "output"
use_4bit = True

# Load dataset
def load_and_preprocess_dataset():
    dataset = load_dataset('json', data_files={'train': train_dataset_path,
                                               'validation': val_dataset_path})  # Assuming the dataset is in JSON format

    # Preprocess function to concatenate context and user_input
    def preprocess_function(examples):
        return {
            'input_texts': [
                f"Context: {context}\nUser Input: {user_input}\nPredict the intent:"
                for context, user_input in zip(examples['context'], examples['user_input'])
            ],
            'labels': examples['intent'],
        }

    # Apply the preprocessing
    tokenized_datasets = dataset.map(preprocess_function, batched=True)
    return tokenized_datasets['train'],tokenized_datasets['validation']



raw_dataset = load_dataset('json', data_files={'train': val_dataset_path})
print(raw_dataset['train'][0])
```

```python
# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(tokenizer_id, use_fast=True)

# Load and preprocess the dataset
train_dataset,validation_dataset = load_and_preprocess_dataset()
```

```python
from transformers import BitsAndBytesConfig

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(
        examples['input_texts'],
        padding="max_length",
        truncation=True,
        max_length=512  # Adjust as per your GPU memory and requirement
    )

train_tokenized_dataset = train_dataset.map(tokenize_function, batched=True)
eval_tokenized_dataset = train_dataset.map(tokenize_function, batched=True)

# Load the Phi-3-mini model
bnb_config = BitsAndBytesConfig(
    load_in_4bit=use_4bit,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16 if torch.cuda.is_bf16_supported() else torch.float16,
    bnb_4bit_use_double_quant=True,
)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    torch_dtype=torch.float16 if torch.cuda.is_bf16_supported() else torch.bfloat16,
    device_map="auto",  # Automatically places the model on the right device (GPU or CPU)
)

# Prepare LoRA configuration for the model
peft_config = LoraConfig(
    r=16,  # Low rank adaptation rank
    lora_alpha=16,  # Alpha scaling factor
    lora_dropout=0.05,  # Dropout for regularization
    target_modules=['q_proj', 'k_proj', 'v_proj', 'o_proj'],  # Target modules in the transformer architecture
    task_type=TaskType.CAUSAL_LM  # Causal language modeling task type
)

# Apply PEFT to the model with LoRA
model = get_peft_model(model, peft_config)

# Training arguments
training_args = TrainingArguments(
    output_dir=output_dir,  # Directory to save the model
    eval_strategy="steps",  # Evaluate the model at regular steps
    eval_steps=100,  # Perform evaluation every 100 steps
    save_strategy="steps",  # Save the model at regular steps
    save_steps=500,
    logging_steps=100,  # Log every 100 steps
    per_device_train_batch_size=4,  # Adjust this based on GPU memory
    gradient_accumulation_steps=4,
    learning_rate=1e-4,  # Learning rate for fine-tuning
    num_train_epochs=3,  # Number of epochs to train
    weight_decay=0.01,  # Weight decay for optimization
    logging_dir='/content/drive/MyDrive/intent-classifier/logs',  # Logging directory
    fp16=True,  # Use mixed precision for faster training
    push_to_hub=False,  # Set to True if you want to push the model to Hugging Face hub
    report_to="none",  # Set this to "wandb" if you want to log the training metrics to Weights and Biases
)

# Define a data collator for sequence-to-sequence tasks
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
train_tokenized_dataset = train_tokenized_dataset.filter(lambda x: len(x['labels']) > 0 and isinstance(x['labels'], list) and len(x['labels'][0]) > 0)
eval_tokenized_dataset = eval_tokenized_dataset.filter(lambda x: len(x['labels']) > 0 and isinstance(x['labels'], list) and len(x['labels'][0]) > 0)
# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_tokenized_dataset,
    eval_dataset=eval_tokenized_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
)
```
